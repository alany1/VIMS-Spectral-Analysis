```python
def filter_data(data,parameter_bounds):
    """
    For each spectra s in data, does not include s in the filtered data if parameters are outside of reasonable range. 
    
    parameter_bounds: dictionary mapping each channel to parameter bounds.
    """
    filtered_data = []
    num_filtered = 0
    num_bad_fits = 0
    noisy_spectra = []
    bad_spectra = []
    sample = False
    
    for i,spectra in enumerate(data):
        if np.max(spectra['spectrum']) < .01:
            noisy_spectra.append(spectra)
            print(f"{len(noisy_spectra)} noisy spectra.")
            continue
        remove = False
        for channel in parameter_bounds:
            if channel not in spectra:
                num_bad_fits+=1
                break
            for parameter in parameter_bounds[channel]:
                if parameter_bounds[channel][parameter][0] < spectra[channel][parameter] < parameter_bounds[channel][parameter][1]:
                    continue
                else:
                    remove = True
                    num_filtered+=1
                    if not sample: print(spectra); sample=True
                    break

            if remove:
                bad_spectra.append(spectra)
                break
                
        if i%10000 == 0: print(f"{num_filtered} filtered and {num_bad_fits} bad fits out of {i+1} spectra.")
            
        if remove:
            continue
        else:
            filtered_data.append(spectra)
            
    return filtered_data, noisy_spectra,bad_spectra

bounds = {'1.0': {'g_amp':(0,.15),'g_mean':(.9,1.1)}, '2.0':{'g_amp':(0,.15),'g_mean':(1.9,2.1)},\
          '1.6':{'g_amp':(0,.15),'g_mean':(1.5,1.7)}}

filtered_data,noisy_spectra,bad_spectra = filter_data(data,bounds)

#better fitting routine (function, empirical corrections from paper), fix bounds , dont fit data that is noise ,
#track other values
#go through all incidence/emission bins and average slopes for each window
#get all rainbows to agree
#use astropy instead of ettime
# from astropy.time import Time 
# tstr = '2013-07-25 12:04:16.735' 
# jd = Time(tstr,format='iso').jd
# print(jd)

```

```python
def time_bucket(data, num_buckets = 20):
    """
    Data: list of spectra dictionaries
    
    Returns: Returns num_buckets lists containing spectra from each time bucket.
    """
    #REASONABLE_AMPLITUDES = (0,0.1)
    
    min_time, max_time = min(data, key = lambda x: x['ettime'])['ettime'],max(data, key = lambda x: x['ettime'])['ettime']
    
    bucket_length = (max_time - min_time) / num_buckets
    
    low = min_time
    up = max_time
    
    buckets = {}
    
    
    for i in range(num_buckets):
        up = bucket_length+low
        buckets[(low,up)] = []
        low = up
    
    
    for i,spec in enumerate(data):
        #if i%10000 == 0: print('Progress', i+1)
        for low,high in buckets:
            #insert cuts here if needed
            if spec['ettime'] <= high and spec['ettime'] >= low:
                buckets[(low,high)].append(spec)
                break
                
    return buckets
        
```

```python
def avg_bucket_spectra(buckets):
    """
    buckets: Dictionary mapping (low,high) interval to list containing spectra.
    
    Returns: Dictionary with one spectra for each interval representing computed direct avg.
    """
    averages = {}
    
    for time in buckets:
        total = np.zeros(len(buckets[time][0]['spectrum']))
        for spectra in buckets[time]:
            total+=spectra['spectrum']
        averages[time] = total/len(buckets[time])
    return averages
```

```python
def plot_buckets(avg_spectra, channel='2.0'):
    """
    avg_spectra: Directly computed average spectra.
    channel: channel number to be examined
    
    Fits avg spectra and then plots spectra for each time bucket.
    """
    REASONABLE_AMPLITUDES = (0,0.1)
    ncolors = len(avg_spectra)
    colors = plt.cm.nipy_spectral(np.linspace(0.05,0.95,ncolors)) 
    
    bad = 0
    for i,bucket in enumerate(avg_spectra):
        fits = fit(avg_spectra[bucket])
        if fits != {}:
            print(fits)
            if fits[channel]['g_amp'] > REASONABLE_AMPLITUDES[1] or fits[channel]['g_amp'] < REASONABLE_AMPLITUDES[0]:
                bad+=1
                break
            mean,sigma,amp = fits[channel]['g_mean'],fits[channel]['g_sigma'],fits[channel]['g_amp']
            x=np.linspace(0,5,1000)
            plt.plot(x,gaussian(x,mean,sigma,amp),color=colors[i],label=bucket)
        else:
            bad += 1
    plt.legend()
    print(f"Bad {bad} out of total {len(avg_spectra)}")

plot_buckets(avg_spectra)
    
```

```python
def fit(info):
    """
    info: spectrum list.
    
    Returns: Parameters of fitting routine as a dictionary of dictionaries.
    """
    band_channels = [29,30,31,32,47,48,49,50,51,52,53,54,55,84,85,86,87,88,89,90]# + [i for i in range(160,181)] 
    plaw = False
    
    fits = {'1.0':{},'1.2':{},'1.6':{},'2.0':{}}
    
    #attempt to fit powerlaw:
    try:
        p_fit,p_cov = sco.curve_fit(powerlaw,vims_wave[band_channels],info[band_channels],p0=(0.005,-2,0),maxfev=100000)
        p=powerlaw(vims_wave,*p_fit)
        adjusted = info-p
        plaw=True
    except:
        adjusted = info
    
    #2.0:
    try:
        fit,cov=sco.curve_fit(gaussian,vims_wave[60:85],adjusted[60:85],p0=(2.0,.125,.1),maxfev=100000)
        if plaw:
            fits['2.0']={'p_amp': p_fit[0],'p_power': p_fit[1],'p_const': p_fit[2],'g_mean':fit[0],'g_sigma': fit[1],'g_amp':fit[2]} 
        else:
            fits['2.0']={'g_mean':fit[0],'g_sigma': fit[1],'g_amp':fit[2]} 
    except:
        return {}
    
    #1.6:
    try:
        fit,cov=sco.curve_fit(gaussian,vims_wave[35:54],adjusted[35:54],p0=(1.6,.125,.1),maxfev=100000)
        if plaw:
            fits['1.6'] = {'p_amp': p_fit[0],'p_power': p_fit[1],'p_const': p_fit[2],'g_mean':fit[0],'g_sigma': fit[1],'g_amp':fit[2]}
        else:
            fits['1.6']={'g_mean':fit[0],'g_sigma': fit[1],'g_amp':fit[2]} 
    except:
        return {}
    
    #1.2:
    try:
        fit,cov=sco.curve_fit(gaussian,vims_wave[12:30],adjusted[12:30],p0=(1.2,.125,.1),maxfev=100000)
        if plaw:
            fits['1.2'] = {'p_amp': p_fit[0],'p_power': p_fit[1],'p_const': p_fit[2],'g_mean':fit[0],'g_sigma': fit[1],'g_amp':fit[2]}
        else:
            fits['1.2']={'g_mean':fit[0],'g_sigma': fit[1],'g_amp':fit[2]} 
    except:
        return {}
    
    try:
        fit,cov=sco.curve_fit(gaussian,vims_wave[:18],adjusted[:18],p0=(1.0,.125,.1),maxfev=100000)
        if plaw:
            fits['1.0'] = {'p_amp': p_fit[0],'p_power': p_fit[1],'p_const': p_fit[2],'g_mean':fit[0],'g_sigma': fit[1],'g_amp':fit[2]}
        else:
            fits['1.0']={'g_mean':fit[0],'g_sigma': fit[1],'g_amp':fit[2]} 
    except:
        return {}
    
    fits['2.0']['channel']=info[69]
    fits['1.0']['channel']=info[8]
    fits['1.2']['channel']=info[20]
    fits['1.6']['channel']=info[45]
    
    return fits
```

```python
def V(x, A, mean, alpha, gamma):
    """
    Return the Voigt line shape at x with Lorentzian component HWHM gamma
    and Gaussian component HWHM alpha.

    """
    sigma = alpha / np.sqrt(2 * np.log(2))

    return np.real(A*wofz(((x-mean) + 1j*gamma)/sigma/np.sqrt(2))) / sigma\
                                                           /np.sqrt(2*np.pi)

def powerlaw(x,a,b):
    return a*x**b
```
